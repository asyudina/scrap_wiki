```python
#import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
from urllib.request import urlopen 
```

```python
#links to scrap companies
s='''/wiki/Category:Dental_companies_of_the_United_States',
 '/wiki/Category:Health_clubs_in_the_United_States',
 '/wiki/Category:Hospital_networks_in_the_United_States',
 '/wiki/Category:Medical_outsourcing_companies_of_the_United_States',
 '/wiki/Category:Members_of_Blue_Cross_Blue_Shield_Association',
 '/wiki/Category:Pharmaceutical_companies_of_the_United_States',
 '/wiki/Category:Defunct_health_care_companies_of_the_United_States','''.split(',\n ')
```

```python
#get links to companies pages
company_urls=[]

for site in s:
    site = site.replace("'", '')
    url = 'https://en.wikipedia.org'+site
    try: 
        page = urlopen(url)
        soup = BeautifulSoup(page, 'html.parser')
        for link in soup.find_all('a'):
            s = link.get('href')
            try:
                if '/wiki/' in s and link.parent.parent.parent.get('class') == ['mw-category-group']: 
                    
                    company_urls.append(s)
            except: 
                continue
    except: 
        continue
```

```python
#get domains (official sites) of companies
domains = []
for site in company_urls:
    lnks = []
    site = site.replace("'", '')
    url = 'https://en.wikipedia.org'+site
    page = urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')
    for link in soup.find_all('span'):
        if link.get('class') == ["official-website"]:
            
            domain = link.a['href']
            domains.append(domain)
            break
        else:  
            domain = 0        
            
    if domain == 0 :

        for link in soup.find_all('span'):
            if link.get('class') == ["url"]:
                domain = link.a['href']
                if 'http://' in domain or 'https://' in domain:
                    lnks.append(domain)


        try:
            domains.append(lnks[0])

        except: 
            for link in soup.find_all('a'):
                if link.get('class') == ['external', 'free']:
                    domain = link['href']
                    if 'http://' in domain or 'https://' in domain:
                        lnks.append(domain)
            try:
                for el in lnks:
                    if len(el)>45:
                        lnks.remove(el)
                domains.append(lnks[0])     
            except: 
                for link in soup.find_all('a'):
                    if link.get('class') == ['external', 'text']:
                        domain = link['href']
                        if 'http://' in domain or 'https://' in domain:
                            lnks.append(domain)
                try:
                    for el in lnks:
                        if len(el)>45:
                            lnks.remove(el)
                    domains.append(lnks[0]) 
                except:
                    domains.append(0) 
```

```python
#get categories of companies listed in wiki (dental/pharmacy/...etc)
categs = {}
for site in company_urls:
    lnks = []
    site = site.replace("'", '')
    url = 'https://en.wikipedia.org'+site
    try:
        page = urlopen(url)
        soup = BeautifulSoup(page, 'html.parser')
        cats = []
        for link in soup.find_all('a'):
            s = link.get('href')
            try:
                if '/wiki/Category:' in s: 
                    if link.get('title').startswith('Category:'):
                        cats.append(link.get('title'))


            except: 
                continue
        categs[site] = cats
    except:
        continue
```

```python
#categories to dataframe
f = pd.DataFrame.from_dict(categs, orient='index')
f.head()
```
